## 作業目標

* 請利用 API: https://www.dcard.tw/_api/forums/pet/posts?popular=true 回答下列問題：

1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

```python
import requests
import json
html = requests.get("https://www.dcard.tw/_api/forums/pet/posts?popular=true")
dataset = json.loads(html.text)
```

```python
dataset
```

```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
'''
Your Code
'''
print(f"1. 這個 API 一次會回傳 {len(dataset)} 筆資料")
print()
print(f"2. 每一筆資料包含 {list(dataset[0].keys())} 欄位")
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
'''
Your Code
'''
for data in dataset:
    print(f'標題: {data["title"]} ')
    print(f'貼文時間: {data["createdAt"]} ')
    print(f'留言人數: {data["commentCount"]} ')
    print(f'按讚人數: {data["likeCount"]} ')
    print()
```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」
'''
Your Code
'''
hot = requests.get("https://www.dcard.tw/_api/forums/pet/posts?popular=true")
hot_dataset = json.loads(hot.text)

not_hot = requests.get("https://www.dcard.tw/_api/forums/pet/posts?popular=false")
not_hot_dataset = json.loads(not_hot.text)

def main_commentcount(dataset):
    commentcount = 0
    for data in dataset:
        commentcount += data["commentCount"]
    return commentcount / len(dataset)

def main_likecount(dataset):
    likecount = 0
    for data in dataset:
        likecount += data["likeCount"]
    return likecount / len(dataset)
    
hot_main_commentcount = main_commentcount(hot_dataset)
not_hot_main_commentcount = main_commentcount(not_hot_dataset)
hot_main_likecount = main_likecount(hot_dataset)
not_hot_main_likecount = main_likecount(not_hot_dataset)

print(f"熱門文章 平均留言人數 {hot_main_commentcount}")
print(f"非熱門文章 平均留言人數 {not_hot_main_commentcount}")
print(f"熱門文章 平均按讚人數 {hot_main_likecount}")
print(f"非熱門文章 平按讚言人數 {not_hot_main_likecount}")
```
