### 根據範例檔案的結果：

```python
f = open("./data/64_72hr_CH.xml", "r", encoding="utf-8")
xml = f.read()
f.close()
```

```python
import xmltodict
d = dict(xmltodict.parse(xml))
```

```python
# 1. 請問高雄市有多少地區有溫度資料？

'''
Your Code
'''
locations = d['cwbopendata']['dataset']['locations']['location']
len(locations)
```

```python
# 2. 請取出每一個地區所記錄的第一個時間點跟溫度

'''
Your Code
'''
for location in locations:
    print(location['locationName'])
    print(location['weatherElement'][0]['time'][0]['dataTime'])
    print(location['weatherElement'][0]['time'][0]['elementValue']['value'], location['weatherElement'][0]['time'][0]['elementValue']['measures'])
    print()
```

```python
# 3. 請取出第一個地區所記錄的每一個時間點跟溫度

'''
Your Code
'''
location_times = locations[0]['weatherElement'][0]['time']

print(locations[0]['locationName'])
for location_time in location_times:
    print(location_time['dataTime'])
    print(location_time['elementValue']['value'], location_time['elementValue']['measures'])
    print()
```
